## Escena 1 — Hook

A veces usamos listas para guardar todo, pero no siempre son la mejor opción.
Si tú ya sabes exactamente qué dato quieres buscar, un diccionario es mucho más cómodo porque te deja ir **directamente por una llave**.

Piensa en esto como una libreta con etiquetas:
- `nombre` → valor
- `edad` → valor
- `ciudad` → valor

Ese es el poder de los diccionarios: organizar información con nombres claros.

```python
persona = {
    "nombre": "Juan",
    "edad": 25,
    "ciudad": "CDMX"
}

print(persona)
print(persona["nombre"])
```


````python
persona = {
    "nombre": "Juan",
    "edad": 25,
    "ciudad": "CDMX"
}

print(persona)
print(persona["nombre"])
````

## Escena 2 — Presentación del problema

Un diccionario sirve para guardar datos relacionados entre sí.
En lugar de recordar posiciones como en una lista, aquí usas palabras clave.

Eso hace que el código sea más fácil de leer.
También ayuda mucho cuando manejas información de usuarios, productos, pedidos o configuraciones.

En este ejemplo guardamos datos de un usuario con llaves que tienen sentido.

```python
usuario = {
    "nombre": "Ana",
    "edad": 30,
    "activo": True
}

print(usuario["nombre"])
print(usuario["activo"])
```


````python
usuario = {
    "nombre": "Ana",
    "edad": 30,
    "activo": True
}

print(usuario["nombre"])
print(usuario["activo"])
````

## Escena 3 — Explicación + código principal

Ahora vamos a un ejemplo más real.
Aquí no guardamos un solo usuario, sino varios.

Cada persona tiene su nombre como llave principal, y adentro guardamos otro diccionario con más datos.
Esto es muy útil porque te permite trabajar con información organizada y acceder rápido a lo que necesitas.

También verás tres cosas importantes:
- cómo acceder a un dato específico,
- cómo agregar un nuevo elemento,
- y cómo recorrer todo el diccionario.

```python
usuarios = {
    "juan": {"edad": 25, "ciudad": "CDMX"},
    "ana": {"edad": 30, "ciudad": "Madrid"}
}

print(usuarios["juan"]["edad"])

usuarios["luis"] = {"edad": 28, "ciudad": "Bogotá"}

for nombre, datos in usuarios.items():
    print(nombre, datos["ciudad"])
```


````python
usuarios = {
    "juan": {"edad": 25, "ciudad": "CDMX"},
    "ana": {"edad": 30, "ciudad": "Madrid"}
}

print("Edad de Juan:", usuarios["juan"]["edad"])

usuarios["luis"] = {"edad": 28, "ciudad": "Bogotá"}

for nombre, datos in usuarios.items():
    print(nombre, "vive en", datos["ciudad"])
````

## Escena 4 — Nivel medio: herramientas prácticas

En el nivel medio no solo importa acceder a datos.
También importa trabajar de forma cómoda y segura.

Por eso conviene conocer estas herramientas:
- `keys()` para ver las llaves,
- `values()` para ver los valores,
- `items()` para recorrer llave y valor juntos,
- `get()` para evitar errores cuando una llave no existe.

Esto te ayuda a escribir código más limpio y más resistente.

```python
producto = {
    "nombre": "Laptop",
    "precio": 15000,
    "stock": 8
}

print(producto.keys())
print(producto.values())
print(producto.get("marca", "No disponible"))

for clave, valor in producto.items():
    print(clave, valor)
```


````python
producto = {
    "nombre": "Laptop",
    "precio": 15000,
    "stock": 8
}

print("Llaves:", producto.keys())
print("Valores:", producto.values())
print("Marca:", producto.get("marca", "No disponible"))

for clave, valor in producto.items():
    print(clave, ":", valor)
````

## Escena 5 — Cierre y siguiente paso

Un error muy común es intentar leer una llave que no existe.
Cuando haces eso con corchetes, Python puede lanzar un error.

La forma más segura en muchos casos es usar `get()`.
Así puedes devolver un valor por defecto y tu programa sigue funcionando.

Este pequeño detalle separa un código frágil de uno más profesional.

```python
usuarios = {
    "juan": {"edad": 25},
    "ana": {"edad": 30}
}

print(usuarios.get("pedro", "No existe ese usuario"))
print(usuarios.get("juan"))
```

Siguiente paso natural: en el nivel difícil, los diccionarios se usan para resolver problemas reales como contar elementos, agrupar datos y transformar estructuras.


````python
usuarios = {
    "juan": {"edad": 25},
    "ana": {"edad": 30}
}

print(usuarios.get("pedro", "No existe ese usuario"))
print(usuarios.get("juan"))
````
