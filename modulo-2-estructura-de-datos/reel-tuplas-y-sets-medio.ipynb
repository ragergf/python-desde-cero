## Escena 1 — Hook: no todo debe ser lista

En Python muchas personas empiezan usando listas para todo, porque son muy fáciles de crear y modificar.

Pero no siempre son la mejor opción.

A veces necesitas una estructura que **no cambie**, y otras veces necesitas una que **evite duplicados automáticamente**.

Aquí vamos a comparar una lista, una tupla y un set para que la diferencia se vea clara desde el inicio.

```python
# Lista: se puede modificar
frutas_lista = ["manzana", "pera", "uva"]
# Tupla: pensada para datos que no deberían cambiar
frutas_tupla = ("manzana", "pera", "uva")
# Set: guarda valores únicos
frutas_set = {"manzana", "pera", "uva", "uva"}
print("Lista:", frutas_lista)
print("Tupla:", frutas_tupla)
print("Set:", frutas_set)
```

## Escena 2 — Cuándo usar tuplas y sets

Una **tupla** se usa cuando quieres guardar datos en un orden fijo y no quieres que cambien fácilmente.

Un **set** se usa cuando te importa tener elementos únicos, es decir, sin repetidos.

Piensa en esto:
- una tupla puede servir para coordenadas, fechas o configuraciones fijas,
- un set puede servir para limpiar datos repetidos, como correos, nombres o IDs.

Vamos a ver cada uno con ejemplos sencillos.

```python

coordenadas = (10, 20)
emails = {"ana@test.com", "luis@test.com", "ana@test.com"}

print("Coordenadas:", coordenadas)
print("Emails únicos:", emails)
```

## Escena 3 — Tuplas: datos que deben mantenerse estables

Una tupla se parece mucho a una lista, pero tiene una diferencia importante:
**no se puede modificar después de crearla**.

Eso la hace útil cuando quieres representar información que debería mantenerse igual, por ejemplo:
- una coordenada,
- una fecha,
- un color RGB,
- o una configuración fija.

No significa que sea “mejor” que una lista.  
Significa que tiene un propósito diferente.

```python

coordenadas = (10, 20)
print("Tupla original:", coordenadas)
print("Primer valor:", coordenadas[0])
print("Segundo valor:", coordenadas[1])
# Podemos recorrerla igual que otras estructuras
for valor in coordenadas:
    print("Valor:", valor)
```

### Importante

Si intentas cambiar una tupla, Python marca error porque fue diseñada para mantenerse estable.

En el reel mostramos ese comportamiento porque ayuda a entender por qué la tupla transmite la idea de “esto no debe cambiar”.

```python

coordenadas = (10, 20)

try:
    coordenadas[0] = 50
except TypeError as e:
    print("Error al modificar la tupla:", e)
```

## Escena 4 — Sets: una forma rápida de eliminar duplicados

Los sets son colecciones de elementos únicos.

Eso significa que si agregas el mismo valor varias veces, Python solo guarda una copia.

Esto es muy útil cuando trabajas con listas que traen datos repetidos y quieres limpiarlos de manera rápida y fácil.

```python

numeros = {1, 2, 2, 3, 3, 3, 4}
print("Set de números:", numeros)
nombres = ["Ana", "Luis", "Ana", "Pedro", "Luis"]
nombres_unicos = set(nombres)
print("Lista original:", nombres)
print("Sin duplicados:", nombres_unicos)
```

### Ojo con esto

Los sets no garantizan el mismo orden visual que una lista o una tupla.

Su objetivo principal no es mantener orden, sino ayudarte a trabajar con **valores únicos** y hacer ciertas operaciones de forma eficiente.

```python

colores = ["rojo", "azul", "rojo", "verde", "azul"]
colores_unicos = set(colores)

print("Lista original:", colores)
print("Set resultante:", colores_unicos)
```

## Escena 5 — Nivel medio: uso real con correos repetidos

Aquí está una aplicación muy real.

Imagina que recibes una lista de correos desde un formulario, un archivo o una API.
Si quieres saber cuántos correos distintos hay, un set te ayuda muchísimo.

Este tipo de lógica aparece en backend, validaciones y procesamiento de datos.

```python
emails = [
    "a@test.com",
    "b@test.com",
    "a@test.com",
    "c@test.com"
]
emails_unicos = set(emails)
print("Lista original:", emails)
print("Emails únicos:", emails_unicos)
print("Cantidad original:", len(emails))
print("Cantidad única:", len(emails_unicos))
```

## Escena 6 — Comparación final rápida

Ya con esto puedes quedarte con una regla mental muy simple:

- **Lista**: cuando necesitas cambiar, agregar o quitar elementos.
- **Tupla**: cuando quieres datos estables.
- **Set**: cuando necesitas eliminar repetidos o trabajar con valores únicos.

Esa pequeña decisión mejora mucho la claridad de tu código.

```python
lista = [1, 2, 2, 3]
tupla = (1, 2, 2, 3)
set_datos = {1, 2, 2, 3}

print("Lista:", lista)
print("Tupla:", tupla)
print("Set:", set_datos)
```
