# Diccionarios en Python: más poder del que parece

Los diccionarios en Python parecen simples al principio, porque solo guardan datos con una clave y un valor.

Pero cuando empiezas a usarlos mejor, te das cuenta de que son una de las herramientas más importantes del lenguaje.

Se usan en APIs, backend, automatización y manejo de datos.

```python
persona = {
    "nombre": "Ana",
    "edad": 28,
    "ciudad": "CDMX"
}

print(persona)
print(type(persona))
```

# Cómo leer información dentro de un diccionario

Para obtener un dato dentro de un diccionario, usamos su clave.

Cada clave apunta directamente a un valor, lo que hace que acceder a la información sea rápido y preciso.

```python
persona = {
    "nombre": "Ana",
    "edad": 28,
    "ciudad": "CDMX"
}

print(persona["nombre"])
print(persona["edad"])
print(persona["ciudad"])
```

# Usar get() para evitar errores

A veces intentas acceder a una clave que no existe.

Si usas la forma tradicional, Python puede lanzar un error.

El método `get()` permite acceder de forma segura y además puedes definir un valor por defecto si la clave no existe.

```python
persona = {
    "nombre": "Ana",
    "edad": 28
}

print(persona.get("nombre"))
print(persona.get("ciudad"))
print(persona.get("ciudad", "No disponible"))
```

# Ver claves, valores y recorrer un diccionario

En nivel medio ya no solo accedes a un dato.

Ahora necesitas ver todo el contenido del diccionario.

Python te permite obtener:
- Todas las claves
- Todos los valores
- Ambos al mismo tiempo

Y recorrerlos fácilmente.

```python
producto = {
    "nombre": "Laptop",
    "precio": 15000,
    "stock": 8
}

print(producto.keys())
print(producto.values())
print(producto.items())

for clave, valor in producto.items():
    print(clave, "->", valor)
```

# Aplicación práctica de diccionarios

Ahora usamos todo lo aprendido en un ejemplo más real.

Los diccionarios permiten:
- Consultar datos
- Manejar información incompleta
- Recorrer datos fácilmente

Esto es muy común en aplicaciones reales.

```python
usuario = {
    "nombre": "Carlos",
    "email": "carlos@email.com",
    "activo": True
}

print("Nombre:", usuario.get("nombre"))
print("Teléfono:", usuario.get("telefono", "No registrado"))

for clave, valor in usuario.items():
    print(f"{clave}: {valor}")
```
