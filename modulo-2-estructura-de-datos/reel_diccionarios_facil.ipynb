## Celda 2 — PRESENTACIÓN

## Celda 3 — EXPLICACIÓN + CÓDIGO

Esto es donde Python se vuelve poderoso.

Aquí no buscas por número… buscas por significado.

“nombre” → Ana

“edad” → 25

Y lo mejor… puedes agregar, cambiar y recorrer todo fácilmente.

Esto ya se parece al mundo real.


```python
persona = {
    "nombre": "Ana",
    "edad": 25,
    "ciudad": "CDMX"
}

# Acceder
print(persona["nombre"])

# Agregar
persona["profesion"] = "Ingeniera"

# Modificar
persona["edad"] = 26

# Recorrer
for clave, valor in persona.items():
    print(clave, ":", valor)
```

## Celda 4 — GIRO MENTAL

Esto ya no es teoría…

Esto es exactamente lo que regresa una API.

Todo en tecnología moderna… son diccionarios dentro de diccionarios.


```python
# Simulación tipo API (diccionario realista)
response = {
    "status": "success",
    "data": {
        "user": "Ana",
        "role": "admin"
    }
}

print(response["data"]["user"])
```
