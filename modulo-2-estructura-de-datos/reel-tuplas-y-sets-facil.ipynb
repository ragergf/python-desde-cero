## Tuplas y sets: dos ideas muy útiles

Una **tupla** se parece a una lista, pero tiene una diferencia importante: **no se puede modificar** después de crearla.

Un **set** también guarda datos, pero su superpoder es este: **elimina elementos repetidos automáticamente**.

Eso hace que cada estructura tenga un uso distinto:
- **Tupla**: cuando quieres datos fijos.
- **Set**: cuando quieres datos únicos, sin duplicados.

```python

lista = ["Ana", "Luis", "Ana", "Pedro"]
tupla = ("latitud", "longitud")
set_nombres = {"Ana", "Luis", "Ana", "Pedro"}
print("Lista:", lista)
print("Tupla:", tupla)
print("Set:", set_nombres)
print("Observa que:")
print("- la lista conserva repetidos")
print("- la tupla guarda datos fijos")
print("- el set quita duplicados automáticamente")
```

## Tupla = datos que no deben cambiar

Piensa en una tupla como una etiqueta cerrada.

Sirve cuando los datos ya están definidos y **deben mantenerse iguales**, por ejemplo:
- coordenadas,
- días de la semana,
- configuraciones fijas,
- valores que no quieres cambiar por error.

Se escribe con paréntesis: `()`.

```python
# Explicación con código: tupla

coordenadas = (19.4326, -99.1332)
print("Coordenadas:", coordenadas)
print("Primer valor:", coordenadas[0])

print("\nUna tupla se puede leer, pero no modificar.")

# Si quitas el comentario de abajo, Python marcará error.
# coordenadas[0] = 20.0000
```

## Set = datos sin repetir

Un set es perfecto cuando te interesa quedarte solo con elementos únicos.

Ejemplos comunes:
- limpiar nombres repetidos,
- saber qué categorías existen,
- comparar grupos,
- evitar duplicados sin hacer mucho trabajo manual.

Se escribe con llaves: `{}`.

```python
# Explicación con código: set

numeros = {1, 2, 2, 3, 3, 3, 4}
print("Set de números:", numeros)

clientes = ["Carlos", "Ana", "Carlos", "Luis", "Ana"]
clientes_unicos = set(clientes)

print("Lista original:", clientes)
print("Clientes únicos:", clientes_unicos)

print("\nEl set eliminó los repetidos automáticamente.")
```

## Piensa como programador pro

La lección no es memorizar nombres.

La lección real es esta: **cada estructura tiene una función**.

Cuando eliges bien, tu código:
- comunica mejor la intención,

Eso ya es empezar a programar con mentalidad profesional.

```python

lista = [10, 20, 20, 30]
tupla = (10, 20, 20, 30)
set_datos = {10, 20, 20, 30}
print("Lista  -> cambia y permite repetidos:", lista)
print("Tupla  -> no cambia y puede tener repetidos:", tupla)
print("Set    -> no guarda repetidos:", set_datos)
print("\nUsa lista cuando quieras modificar.")
print("Usa tupla cuando el dato debe quedarse fijo.")
print("Usa set cuando quieras valores únicos.")
```
