# Tuplas y Sets — Nivel Difícil (Ejemplo real)

Mini notebook alineado al reel de **Tuplas y Sets — nivel difícil**.

Aquí verás un caso más cercano a producción: **detectar eventos duplicados, validar acceso rápido y proteger configuraciones que no deben cambiar**.

## Idea clave
- **Tupla** → datos fijos, seguros, que no deberían cambiar.
- **Set** → valores únicos y búsquedas rápidas.


## Escena 1 — Hook

### Idea teórica
Mucha gente usa listas para todo porque son fáciles de entender.  
El problema es que **no siempre son la mejor herramienta**.

Cuando trabajas con datos reales, como accesos de usuarios, logs o eventos del sistema, elegir mal la estructura puede hacer tu código:
- más lento,
- más confuso,
- y más propenso a errores.

En esta primera parte vamos a ver un ejemplo simple con una lista y luego lo vamos a mejorar.


```python
# Ejemplo inicial: una lista con eventos del sistema
eventos = [
    "LOGIN:ana",
    "LOGIN:juan",
    "LOGIN:ana",
    "RESET:maria",
    "LOGIN:juan"
]

print("Eventos originales:")
for evento in eventos:
    print("-", evento)

print("\nTotal de eventos:", len(eventos))
```

## Escena 2 — El problema

### Idea teórica
Ahora piensa en un caso real: un sistema recibe muchos eventos y necesita responder preguntas como estas:

- ¿este usuario ya apareció?
- ¿hay eventos duplicados?
- ¿esta configuración del sistema se puede modificar?

Aquí es donde entran **sets** y **tuplas**.

Un **set** sirve cuando te importan los valores únicos y quieres buscar rápido.  
Una **tupla** sirve cuando un grupo de datos debe quedarse fijo, como una configuración o una clave compuesta.


```python
# Tupla: configuración fija del sistema
CONFIG_SISTEMA = ("produccion", 443, True)

# Set: usuarios permitidos para entrar a una zona crítica
usuarios_autorizados = {"ana", "maria", "sofia"}

print("Configuración fija:", CONFIG_SISTEMA)
print("Usuarios autorizados:", usuarios_autorizados)

usuario = "ana"
print(f"\n¿{usuario} está autorizado?", usuario in usuarios_autorizados)

usuario = "carlos"
print(f"¿{usuario} está autorizado?", usuario in usuarios_autorizados)
```

## Escena 3 — Explicación práctica con ejemplo real difícil

### Idea teórica
Vamos a usar un caso más serio: imagina un sistema que recibe eventos de seguridad.  
Cada evento tiene:
- usuario,
- acción,
- IP.

Queremos detectar duplicados de manera eficiente.

Aquí la idea profesional es convertir cada evento en una **tupla**, porque cada registro debe representar un dato fijo.  
Luego guardamos esas tuplas en un **set** para eliminar duplicados y validar pertenencia muy rápido.


```python
# Lista de eventos recibidos por el sistema
logs = [
    ("ana", "LOGIN", "10.0.0.1"),
    ("juan", "LOGIN", "10.0.0.2"),
    ("ana", "LOGIN", "10.0.0.1"),   # duplicado
    ("maria", "RESET_PASSWORD", "10.0.0.8"),
    ("juan", "LOGIN", "10.0.0.2"),  # duplicado
    ("sofia", "LOGIN", "10.0.0.5"),
]

print("Logs recibidos:")
for log in logs:
    print(log)

logs_unicos = set(logs)

print("\nLogs únicos:")
for log in logs_unicos:
    print(log)

print("\nTotal recibidos:", len(logs))
print("Total únicos:", len(logs_unicos))
print("Duplicados detectados:", len(logs) - len(logs_unicos))
```

## Escena 4 — Nivel difícil / Mentalidad profesional

### Idea teórica
Aquí está el error común: usar una lista para validar si algo ya existe cuando el sistema crece mucho.

Con pocos datos quizá no lo notes.  
Pero con miles o millones de registros, esa decisión cuesta tiempo.

### Regla mental profesional
- usa **lista** cuando el orden y la flexibilidad importan,
- usa **set** cuando necesitas validar existencia muy rápido,
- usa **tupla** cuando quieres representar datos fijos o seguros.

Vamos a comparar una validación con lista contra una con set.


```python
# Simulación de IDs de eventos ya procesados
ids_procesados_lista = [101, 102, 103, 104, 105, 106, 107]
ids_procesados_set = set(ids_procesados_lista)

nuevo_id = 105

print("Búsqueda en lista:")
if nuevo_id in ids_procesados_lista:
    print(f"El evento {nuevo_id} ya fue procesado")

print("\nBúsqueda en set:")
if nuevo_id in ids_procesados_set:
    print(f"El evento {nuevo_id} ya fue procesado")

# Flujo realista: evitar reprocesar un evento
evento_entrante = ("ana", "TRANSFER", "txn-9001")
procesados = {
    ("ana", "LOGIN", "txn-1001"),
    ("juan", "PAGO", "txn-1002")
}

if evento_entrante in procesados:
    print("\nEvento repetido, no se procesa")
else:
    procesados.add(evento_entrante)
    print("\nEvento nuevo, se procesa y se registra")

print("\nEventos procesados:")
for item in procesados:
    print(item)
```

## Escena 5 — Cierre / Regla final

### Idea teórica
La gran lección no es memorizar definiciones.  
La gran lección es **elegir bien**.

Cuando eliges la estructura correcta:
- el código se entiende mejor,
- el sistema responde mejor,
- y tú empiezas a pensar como alguien que diseña soluciones, no solo como alguien que escribe líneas.

Quédate con esta regla:
- **Lista** → colección flexible.
- **Tupla** → dato fijo.
- **Set** → valor único y búsqueda rápida.


```python
# Resumen final con un mini caso integrado

config_api = ("api.systemhandler.com", 443, "v1")   # tupla: fijo
usuarios_activos = {"ana", "juan", "maria"}         # set: únicos

cola_eventos = [
    ("ana", "LOGIN"),
    ("ana", "LOGIN"),
    ("juan", "PAGO"),
    ("maria", "LOGOUT"),
]

eventos_unicos = set(cola_eventos)

print("Configuración API:", config_api)
print("Usuarios activos:", usuarios_activos)
print("\nEventos originales:", cola_eventos)
print("Eventos únicos:", eventos_unicos)

print("\nRegla mental final:")
print("- Tupla: configuración fija")
print("- Set: sin duplicados y búsqueda rápida")
print("- Lista: secuencia flexible")
```
